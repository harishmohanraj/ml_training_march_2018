```python
import pandas as pd
```

```python
iris_dataframe = pd.read_csv('./classification/dataset/iris.data.csv')
```

```python
iris_dataframe.head()
```

```python
# Select and index data

# Accessing a column

iris_dataframe[['sepal_length']].head()
```

```python
iris_dataframe[['sepal_length', 'species']].tail()
```

```python
# Accessing a Row: using slicing
# dataframe[start_index:end_index] Remember: The end of the slice is not included and the index starts from 0

iris_dataframe[1:3]
```

```python
# Advanced dataframe accessing

# loc - a technique to select parts of your data based on label
# iloc - a technique to select parts of your data based on position
```

```python
# loc - similar to the above example with [] but more powerful

iris_dataframe.loc[:, ['petal_length', 'petal_width', 'species']].head()
```

```python
iris_dataframe.iloc[:, [2,3,4]].head()
```
